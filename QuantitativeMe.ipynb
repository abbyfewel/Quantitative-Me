### What Statistical Hypothesis Tests are Computed

The first hypothesis we will testing is that there is a significant difference between the amount of quizzes given out *per day* in a semester versus the number of tests *per day*. 

$H_{0}$: $\overline{x}_{quizzes}$ $\leq$ $\overline{x}_{tests}$  
$H_{1}$: $\overline{x}_{quizzes}$ $>$ $\overline{x}_{tests}$

The code is implemented in `utils.tests_quizzes_sig()`, and results are printed.

```python
import importlib
import utils
importlib.reload(utils)

utils.tests_quizzes_sig()
```

**Result:**  
We see that there is a significant difference between the number of quizzes taken over the semester and the number of tests taken of the semester. There were significantly more quizzes taken than tests. 

The next hypothesis that we will be testing is if there is significance in the number of STEM assignments done this semester and non-STEM assignments done. This was inspired by the last pie chart above. 

$H_{0}$: $\overline{x}_{STEM}$ $=$ $\overline{x}_{nonSTEM}$  
$H_{1}$: $\overline{x}_{STEM}$ $!=$ $\overline{x}_{nonSTEM}$

```python
import importlib
import utils
importlib.reload(utils)

utils.stem_v_non_stem_sig()
```

**Result:**  
The null hypothesis that the work loads were the same between STEM and non-STEM classes is not rejected, indicating that work load is similar.
